#### Question 2

```python
def R_coef(theta, train, Y):
    Y_h = train@theta
    mean = Y.sum()/len(Y)
    result = (np.linalg.norm(Y_h-Y)**2/np.linalg.norm((Y-mean))**2)
    result = ((np.linalg.norm(Y_h-mean)**2)/np.linalg.norm((Y-mean))**2)
    return result

    
    
def MSE(theta, train, Y):
    Y_h = np.array(train)@theta
    return ((np.linalg.norm(Y_h - Y))**2)/len(Y)


    
```

#### Question 3

```python
print("determination coefficient : " ,R_coef(reg_lin.coef_, X_train, Y_train))
```

```python
print("MSE of the test data : ", MSE(reg_lin.coef_, X_test, Y_test))
```
