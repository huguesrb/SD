```python
def determination_coeff(Y_hat, Y):
    """
    Computes the determination coefficient
    """
    n = len(Y)
    mean = Y.mean()

    R_2 = 1 - (np.linalg.norm(Y_hat-Y)**2/np.linalg.norm((Y-mean))**2)
    R_2_1 = ((np.linalg.norm(Y_hat-mean)**2)/np.linalg.norm((Y-mean))**2)
    return R_2


def mean_squared_error(Y_hat, Y):
    """
    computes the mean squared error
    """
    n = len(Y)
    return ((np.linalg.norm(Y_hat - Y))**2)/n
```

### Q10 : Forward Variable Selection method

```python
n = len(y_train)


def search_k_star(r, X, A):
    T_list = []
    T = 0
    k_star = 0
    X_tild_chosen = X[:, [0]]
    theta_hat_chosen = linear_model.LinearRegression().fit(X_tild_chosen, r).coef_
    for k in A:
        X_tild = X[:, [k]]
        reg = linear_model.LinearRegression().fit(X_tild, r)
        theta_hat = reg.coef_
        s_hat = 1/(np.sum(X[:, k]**2))
        sigma_hat = 1/(n-1) * np.linalg.norm(r - reg.predict(X_tild))**2
        T_k = np.linalg.norm(theta_hat, ord=1)/(np.sqrt(s_hat*sigma_hat))
        if np.linalg.norm(T_k) > np.linalg.norm(T):
            T = T_k
            k_star = k
            X_tild_chosen = X_tild
            theta_hat_chosen = theta_hat
        T_list.append(T_k)

    return k_star, X_tild_chosen, theta_hat_chosen, 2*(1-stats.norm.cdf(T)), T_list


def forward_variable_selection(X, Y):
    Pvalues_List = []
    test_list = []
    n = np.shape(X)[0]
    To_return = []
    r = Y
    A = [i for i in range(np.shape(X)[1])]

    while A != []:
        k_s, X_t_c, theta_h_c, p_v, t = search_k_star(r, X, A)
        del A[A.index(k_s)]
        r = r - X_t_c@theta_h_c
        To_return.append(k_s)
        Pvalues_List.append(p_v)
        test_list.append(t)

    return To_return, Pvalues_List, test_list
```

```python
Selection_List, Pvalues_List, T_list = forward_variable_selection(
    np.array(X_train), y_train)
```
